```python
############################## THIS IS HOMEWORK
############################## THIS IS HOMEWORK
############################## THIS IS HOMEWORK
############################## THIS IS HOMEWORK
############################## THIS IS HOMEWORK
############################## THIS IS HOMEWORK

# create function that pulls files based off directory and first few things
def read_files_that_start_with(path_name, subscript):
    """read_files_that_start_with looks within your path name for any files that start with your inputted subscript
    BOTH INPUTS MUST BE WITHIN ""
    path_name MUST start with a /
    EXAMPLE: read_files_that_start_with("/File/path/here","startswith")
    """
    final_file_name = path_name
    final_file_name += "/"
    final_file_name += subscript
    final_file_name += "*"
    return (glob.glob(final_file_name))
```

```python
# We create ONE list with 3 different columns. Each column relates to each entire file.
list_of_msft_df = []
for i in files_msft:
    j = pd.read_csv(i,infer_datetime_format = True)
    list_of_msft_df.append(j)
```

```python
# Create a list all_dict of 3 different dictionaries we are calling dict_for_df
# As a check, print the length of each dictionary
all_dict = []

# For every data frame in this list of data frames
for df in list_of_msft_df:
    # Create a dictionary for each data frame
    dict_for_df = {}
    # For every row, assign the first column value to date then all the rest of the column values to price
    for row in df.iterrows():
        date = row[1][0]
        price = row[1][1:]
        # Create a dictionary where date is the key and price is the values
        dict_for_df[date] = price
    # Create the list with all 3 dictionaries
    all_dict.append(dict_for_df)
    print("Length of dict_for_df:", len(dict_for_df))

# Create a combined dictionary
combined_dict = {}

# For every dictionary in the list of dictionaries
for each_dict in all_dict:
    # Go through every key and value of the tuple returned by .items()
    for k, v in each_dict.items():
        # And assign these keys and values into our combined dictionary
        combined_dict[k] = v
# Check our combined dictionary length to see if it is indeed the size of adding all 3 dictionaries together
print("Length of combined_dict:", len(combined_dict))
```

```python
# NOW CREATE FILE FROM THE COMBINED FILES. MUCH FASTER TO CREATE A FILE STRAIGHT FROM THE FILES INSTEAD OF
# TRYING TO DO THROUGH DICTIONARIES

# Initialize list
write_list_of_text = []
# go through every file name and add every line into our list. We will have 3 items in the list (for 3 files)
# and every item in that list will have its own list of data
for filename in files_msft:
    with open (filename, 'r') as f:
        write_list_of_text.append(f.readlines())

# Create another list so that we can take every line of text from all three elements in write_list_of_text
# This will make a final list that has every line of text from all 3 files.
combined_final_text = []
for s in write_list_of_text:
    combined_final_text += s[1:] # same as write_list_of_text[0][1:] + write_list_of_text[1][1:] + write_list_of_text[2][1:]
# Here we add the column names
combined_final_text = [write_list_of_text[0][0]] + combined_final_text

# Write file
with open("/Users/Gian/GitHub/Giancarlo-Programming-for-Analytics/Assignments/financial_table.csv", "w") as f:
    f.writelines(combined_final_text)
```
